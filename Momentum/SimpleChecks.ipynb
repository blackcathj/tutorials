# Three station test case

Three station at R = 10, 40, 80 cm, which is the most simple case whose performance is calculable with simple sagita formula.

## Muon with pion fit

```c++
{
    
    TFile *_file0 = TFile::Open("FastTrackingEval_muon.root");
    TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");
    
    TCanvas * c1 = new TCanvas();
    tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_mu(100,.8,1.2)");
    hpT_mu->SetTitle("muon fit;p_{T}/p_{T,truth};count/bin");
    c1 -> Draw();
}
```

## electron with pion fit

```c++
{
    
    TFile *_file0 = TFile::Open("FastTrackingEval_electron_pionfit.root");
    TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");

    TCanvas * c1 = new TCanvas();
    
    tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_e(100,.8,1.2)");
    hpT_e->SetTitle("electron with pion fit;p_{T}/p_{T,truth};count/bin");

    c1 -> Draw();
}
```

## electron with electron fit

As you can see although the peak is moved over, the mean that including the tails is somewhat closer to 1.0

```c++
{
    
    TFile *_file0 = TFile::Open("FastTrackingEval_electron_efit.root");
    TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");

    TCanvas * c1 = new TCanvas();
    
    tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_e(100,.8,1.2)");
    ((TH1*)(gDirectory->Get("hpT_e")))->SetTitle("electron with electron fit;p_{T}/p_{T,truth};count/bin");

    c1 -> Draw();
}
```

## Now overlay multiple cases together

```c++
{
    TCanvas * c1 = new TCanvas();
    
    {
        TFile *_file0 = TFile::Open("FastTrackingEval_electron_efit.root");
        assert(_file0);
        TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");
        assert(tracks);
        
        TH1F * hpT_e = new TH1F("hpT_e","hpT_e",100,.8,1.2);
        tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_e","","goff");
    
        hpT_e->SetLineColor(kBlue+2);
        hpT_e->SetLineWidth(3);
        hpT_e->Draw("");
        hpT_e->SetTitle("electron with electron fit (blue)/pion fit(red) ;p_{T}/p_{T,truth};count/bin");

    }
    
    {
        TFile *_file0 = TFile::Open("FastTrackingEval_electron_pionfit.root");
        assert(_file0);
        TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");
        assert(tracks);
        
        TH1F * hpT_pe_pi = new TH1F("hpT_pe_pi","hpT_pe_pi",100,.8,1.2);
        tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_pe_pi","","goff");
        
        hpT_pe_pi->SetLineColor(kRed-3);
        hpT_pe_pi->SetLineWidth(3);

        hpT_pe_pi->Draw("same");
    }
    
    {
        TFile *_file0 = TFile::Open("FastTrackingEval_thick_electron_efit.root");
        assert(_file0);
        TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");
        assert(tracks);
        
        TH1F * hpT_e = new TH1F("hpT_e","hpT_e",100,.8,1.2);
        tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_e","","goff");
    
        hpT_e->SetLineColor(kBlue+2);

        hpT_e->SetLineStyle(kDashed);
        hpT_e->SetLineWidth(4);        
        hpT_e->Draw("same");
    }
    
    {
        TFile *_file0 = TFile::Open("FastTrackingEval_thick_electron_pionfit.root");
        assert(_file0);
        TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");
        assert(tracks);
        
        TH1F * hpT_pe_pi = new TH1F("hpT_pe_pi","hpT_pe_pi",100,.8,1.2);
        tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_pe_pi","","goff");
        
        hpT_pe_pi->SetLineColor(kRed-3);

        hpT_pe_pi->SetLineStyle(kDashed);
        hpT_pe_pi->SetLineWidth(4);
        hpT_pe_pi->Draw("same");
    }
    
    {
        TFile *_file0 = TFile::Open("FastTrackingEval_thick2_electron_efit.root");
        assert(_file0);
        TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");
        assert(tracks);
        
        TH1F * hpT_e = new TH1F("hpT_e","hpT_e",100,.8,1.2);
        tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_e","","goff");
    
        hpT_e->SetLineColor(kBlue+2);

        hpT_e->SetLineStyle(kDashed);
        hpT_e->SetLineWidth(2);        
        hpT_e->Draw("same");
    }
    
    {
        TFile *_file0 = TFile::Open("FastTrackingEval_thick2_electron_pionfit.root");
        assert(_file0);
        TTree * tracks = (TTree *)_file0->GetObjectChecked("tracks","TTree");
        assert(tracks);
        
        TH1F * hpT_pe_pi = new TH1F("hpT_pe_pi","hpT_pe_pi",100,.8,1.2);
        tracks->Draw("sqrt(px**2 + py**2)/sqrt(gpx**2 + gpy**2)>>hpT_pe_pi","","goff");
        
        hpT_pe_pi->SetLineColor(kRed-3);

        hpT_pe_pi->SetLineStyle(kDashed);
        hpT_pe_pi->SetLineWidth(2);
        hpT_pe_pi->Draw("same");
    }
    
     c1 -> Draw();
}
```
